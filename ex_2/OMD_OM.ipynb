Импорт библиотек

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Считывание файла

```python
data = pd.read_excel("/content/data_stores.xlsx", sheet_name="​Stores")
```

Создаём новый датафрейм, добавляем в него первые 150 строк

```python
new_data = pd.DataFrame({"№ ТТ":data.iloc[1:149, 0], "ГОРОД":data.iloc[1:149, 3],
                         "РЕГИОН":data.iloc[1:149, 4],
                         "ШИР.":data.iloc[1:149, 5], "ДОЛ.":data.iloc[1:149, 6],
                         "ДАТА ОТКР.":data.iloc[1:149, 7],
                         "ДАТА ЗАКР.":data.iloc[1:149, 8]})
pd.set_option('display.max_rows', None)
new_data.head(200)
```

Добавляем оставшиеся строки, учитывая смещение

```python
new_data_second = pd.DataFrame({"№ ТТ":[x for x in range(151, 279)],
                                "ГОРОД":data.iloc[150:, 2],
                                "РЕГИОН":data.iloc[150:, 3],
                                "ШИР.":data.iloc[150:, 4],
                                "ДОЛ.":data.iloc[150:, 5],
                                "ДАТА ОТКР.":data.iloc[150:, 6],
                                "ДАТА ЗАКР.":data.iloc[150:, 7]})
new_data_second.head(200)
```

Сшиваем два датафрейма, сбрасываем индексы

```python
finished_data_Stores = pd.concat([new_data, new_data_second])
finished_data_Stores.reset_index(drop= True, inplace= True)
finished_data_Stores.head(400)
```

Получаем список листов в файле xlx, убираем ненужные

```python
excel_file_path = "/content/data_stores.xlsx"
xls = pd.ExcelFile(excel_file_path)
sheet_names = xls.sheet_names

sheet_names_updated = []
for i in sheet_names:
  if 'Sales' in i:
    sheet_names_updated.append(i)

print(sheet_names_updated)
```

Создаём датафрейм sales, учитывая пустые строки в начале

```python
column_names = ["№ TT", "НЕДЕЛЯ", "КОЛ-ВО"]
sales_data = pd.DataFrame(columns=column_names)
for i in sheet_names_updated:
  data = pd.read_excel("/content/data_stores.xlsx", sheet_name=i, header = None)
  data.columns = column_names

  if pd.isna(data.iloc[0]).any():
    data = data.dropna(axis=0, how='all')
    data.reset_index(drop= True, inplace= True)
    data = data.iloc[1:]
    sales_data = pd.concat([sales_data, data])

  else:
    data = data.iloc[1:]
    sales_data = pd.concat([sales_data, data])

sales_data.reset_index(drop= True, inplace= True)
sales_data.head(len(sales_data))
```

Смотрим № ТТ в различных городах, ищем пересечение

```python
column_name = '№ TT'
for i in sheet_names_updated:
    data = pd.read_excel("/content/data_stores.xlsx", sheet_name=i, header = None)
    data.columns = column_names

    if pd.isna(data.iloc[0]).any():
      data = data.dropna(axis=0, how='all')
      data.reset_index(drop= True, inplace= True)
      data = data.iloc[1:]

    else:
      data = data.iloc[1:]

    unique_values = data[column_name].unique()
    print("Уникальные значения № TT в таблице '{}':".format(i))
    print(unique_values)
```

Создаём объединённый excel файл из таблиц stores и sales

```python
excel_file_path = "/content/OMD_OM_dataset.xlsx"

with pd.ExcelWriter(excel_file_path, engine='openpyxl') as writer:
    finished_data_Stores.to_excel(writer, sheet_name='stores', index=False)

    sales_data.to_excel(writer, sheet_name='sales', index=False)
```

Базовый анализ данных по странице stores

```python
file_path = '/content/OMD_OM_dataset.xlsx'
df = pd.read_excel(file_path, sheet_name='stores')

print(df.head())

# Статистический анализ данных
statistics = df.describe()
print(statistics)

plt.figure(figsize=(10, 6))
sns.countplot(x='ГОРОД', data=df)
plt.title('Количество открытых точек в городах')
plt.xlabel('Город')
plt.ylabel('Количество точек')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Базовый анализ данных по странице sales

```python
df_sales = pd.read_excel(file_path, sheet_name='sales')

# Статистический анализ данных
statistics_sales = df_sales.describe()
print(statistics_sales)

# Визуализация данных
# Гистограмма по количеству продаж для каждого номера TT
plt.figure(figsize=(10, 6))
sns.histplot(x='№ TT', data=df_sales, bins=20)
plt.title('Гистограмма по количеству продаж для каждого номера TT')
plt.xlabel('№ TT')
plt.ylabel('Количество продаж')
plt.grid(True)
plt.show()

# График количества продаж в зависимости от номера TT
plt.figure(figsize=(10, 6))
sns.lineplot(x='№ TT', y='КОЛ-ВО', data=df_sales, marker='o')
plt.title('График количества продаж в зависимости от номера TT')
plt.xlabel('№ TT')
plt.ylabel('Количество продаж')
plt.grid(True)
plt.show()
```
